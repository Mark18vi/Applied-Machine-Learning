```python
## Concat dataframes
import pandas as pd
india_weather = pd.DataFrame({
        "City":['Hyd','Banglore','Delhi'],
        "Temperature":[72,46,91],
        "Wind speed":[20,13,12]
    })
india_weather
```

```python
import pandas as pd
world_weather = pd.DataFrame({
        "City":['NYC','California','Paris','Washington'],
        "Temperature":[70,46,91,33],
        "Wind_speed":[20,13,26,12]
    })
world_weather
```

```python
pd.concat([india_weather,world_weather],ignore_index=True)
```

```python
pd.concat([india_weather,world_weather],axis = 1)
```

## Merge Dataframe

```python
import pandas as pd
india_weather = pd.DataFrame({
        "City":['Hyd','Banglore','Delhi'],
        "Temperature":[72,46,91]
    })
india_weather
```

```python
import pandas as pd
india_humidity = pd.DataFrame({
        "City":['Hyd','Banglore','Delhi'],
        "Temperature":[72,46,91],
        "Humidity":[20,13,12]
    })
india_humidity
```

```python
pd.merge(india_weather,india_humidity,on = 'City')
```

## Numerical Indexing (loc or iloc)

```python
pd.DataFrame([1,3,42,5,42,5,2,5,1,5])
```

```python
df=pd.DataFrame([1,3,42,5,42,5,2,5,1,5],index=[48,49,50,1,2,3,4,5,6,7])
print(df[49])
```

```python
df.loc[49]    # takes index number/ specific index number
```

```python
df.loc[:2]
```

```python
df.iloc[1]   # takes index location or row number
```

```python
df.iloc[:2]
```
