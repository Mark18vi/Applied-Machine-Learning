```python
   matplotlib.pyplot is a collection of command style functions that makes matplot lib as MATLAB
    - takes list (or) array as input
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,3,5,8,3])
plt.ylabel("Numbers")
plt.xlabel('indices')
plt.title("Matplot")

plt.show()
```

```python
plt.plot([1,2,3,4],[2,4,8,16]) # first is x-axis, second is y-axis
plt.ylabel("Almost squares")
plt.xlabel('Numbers')

plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[2,4,8,16],'r*') # first is x-axis, second is y-axis
plt.ylabel("Almost squares")
plt.xlabel('Numbers')

plt.grid()
plt.show()
```

```python
import numpy as np

a = np.arange(0,5,0.2)
 
plt.plot(a, a**2, 'r*', label='^2')
plt.plot(a, a**4, 'g^', label='^4')
plt.plot(a, a**3, 'bs', label='^3')

plt.grid()

plt.legend()
plt.show()
```

```python
x = [1,2,3,4]
y = [2,4,6,8]

plt.plot(x,y, linewidth=3.0)
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [2,4,6,8]
x2 = [4,5,6,7]
y2 = [8,10,12,14]

lines = plt.plot(x1,y1,x2,y2)

# using kw args
plt.setp(lines[0],color='r',linewidth=4.0)
plt.setp(lines[1],'color','b','linewidth','5.0') # MATLAB style

plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3,4])
plt.subplot(212)
plt.plot([-1,-2,-3,-4])

plt.figure(2)
plt.plot([4,5,6])

plt.figure(1)
plt.subplot(211)
plt.title(['Easy',1,2,3])
plt.show()
```
